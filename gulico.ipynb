```
train_x = pcc_tr.drop('pcc-全国-実績',axis=1)
train_y = pcc_tr['pcc-全国-実績']
test_x = pcc_te.drop('pcc-全国-実績',axis=1)

print(train_x.shape,train_y.shape , test_x.shape)
```

```
#1学習データとバリデーションに分ける

from sklearn.model_selection import KFold
kf = KFold(n_splits = 4,shuffle = True)
tr_idx , va_idx = list(kf.split(train_x))[0]
tr_x , va_x = train_x.iloc[tr_idx] , train_x.iloc[va_idx]
tr_y , va_y = train_y.iloc[tr_idx] , train_y.iloc[va_idx]
```

```
#lightGBMを実装する
import lightgbm as lgb 
from sklearn.metrics import mean_absolute_error

#特徴量と目的変数をlightgbmのデータ構造に変換する
lgb_train = lgb.Dataset(tr_x , tr_y) 
lgb_eval = lgb.Dataset(va_x , va_y)

params = {
  'objective':'mae',
  'max_depth':12,
  'alpha': 1,
  'gamma': 1,
  'min_child_weight':1,
  'base_score':7.76
}
```

```
model_fair = lgb.train(params , lgb_train , valid_names=['train','valid'],valid_sets=[lgb_train,lgb_eval])
```

```
# バリデーションデータでのスコアの確認
va_pred = model_fair.predict(va_x)
score = mean_absolute_error(va_y, va_pred)
print(f'MAE: {score:.4f}')
```

```
# 予測
pred = model_fair.predict(test_x)
```
